```python
# import the necessary packages
from matplotlib import pyplot as plt
import argparse
import cv2
```

```python
path = '/content/drive/MyDrive/Plant/Leaf/q_aff_cerris/R_0Class1 (10).jpg'
image = cv2.imread(path)

plt.title('Original')
plt.imshow(image)
plt.show()
```

```python
# images are simply NumPy arrays -- with the origin (0, 0) located at
# the top-left of the image
(b, g, r) = image[0, 0]
print ('Pixel at (0, 0) - Red: {}, Green: {}, Blue: {}'.format(r, g, b))

# access the pixel located at x=50, y=20
(b, g, r) = image[20, 50]
print ('Pixel at (50, 20) - Red: {}, Green: {}, Blue: {}'.format(r, g, b))

# update the pixel at (50, 20) and set it to red
image[20, 50] = (0, 0, 255)
(b, g, r) = image[20, 50]
print ('Pixel at (50, 20) - Red: {}, Green: {}, Blue: {}'.format(r, g, b))
```

```python
(h, w) = image.shape[:2]

(cX, cY) = (w // 2, h // 2)
tl = image[0:cY, 0:cX]
plt.imshow(tl)
plt.title('Top - Left Corner')
```

```python
# in a similar fashion, we can crop the top-right, bottom-right, and
# bottom-left corners of the image and then display them to our
# screen
tr = image[0:cY, cX:w]
plt.imshow(tr)
plt.title("Top-Right Corner")
```

```python
br = image[cY:h, cX:w]
plt.imshow(br)
plt.title('Bottom-Right Corner')
```

```python
bl = image[cY:h, 0:cX]
plt.imshow(bl)
plt.title('Bottom-Left Corner')
```

```python
# set the top-left corner of the original image to be green
image[0:cY, 0:cX] = (0, 255, 0)

# Show our updated image
plt.imshow(image)
plt.title('Updated')
```
