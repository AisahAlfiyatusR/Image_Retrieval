```python
# import the necessary packages
from matplotlib import pyplot as plt
import numpy as np
import argparse
import cv2
```

```python
path = '/content/drive/MyDrive/Plant/Leaf/q_aff_cerris/R_0Class1 (10).jpg'
image = cv2.imread(path)
plt.title('Original')
plt.imshow(image)
plt.show()

print('[INFO] Flipping Image Horizontally.....')
flipped = cv2.flip(image, 1)
plt.imshow(flipped)
plt.title('Flipped Horizontally')
```

```python
print('[INFO] Flipping Image Vertically.....')
flipped = cv2.flip(image, 0)
plt.imshow(flipped)
plt.title('Flipped Vertically')
```

```python
print('[INFO] Flipping Image Horizontally and Vertically.....')
flipped = cv2.flip(image, -1)
plt.imshow(flipped)
plt.title('Flipped Horizontally and Vertically')
```
