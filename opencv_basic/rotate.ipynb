```python
# import the necessary packages
import matplotlib.pyplot as plt
import argparse
import imutils
import cv2
```

```python
path = '/content/drive/MyDrive/Plant/Leaf/q_aff_cerris/R_0Class1 (10).jpg'
image = cv2.imread(path)

plt.title('Original')
plt.imshow(image)
plt.show()
```

```python
# grab the dimensions of the image and calculate the center of the
# image
(h, w) = image.shape[:2]
(cX, cY) = (w // 2, h // 2)

# rotate our image by 45 degrees around the center of the image
M = cv2.getRotationMatrix2D((cX, cY), 45, 1.0)
rotated = cv2.warpAffine(image, M, (w, h))
plt.imshow(rotated)
plt.title("Rotated by 45 Degrees")
```

```python
# rotate our image by -90 degrees around the image
M = cv2.getRotationMatrix2D((cX, cY), -90, 1.0)
rotated = cv2.warpAffine(image, M, (w, h))
plt.imshow(rotated)
plt.title("Rotated by -90 Degrees")
```

```python
# rotate our image around an arbitrary point rather than the center
M = cv2.getRotationMatrix2D((10, 10), 45, 1.0)
rotated = cv2.warpAffine(image, M, (w, h))
plt.imshow(rotated)
plt.title("Rotated by Arbitrary Point")
```

```python
# use our imutils function to rotate an image 180 degrees
rotated = imutils.rotate(image, 180)
plt.imshow(rotated)
plt.title("Rotated by 180 Degrees")
```

```python
# rotate our image by 33 degrees counterclockwise, ensuring the
# entire rotated image still views in the viewing area
rotated = imutils.rotate_bound(image, -33)
plt.imshow(rotated)
plt.title("Rotated Without Cropping")
```
