```python
# import the necessary packages
from matplotlib import pyplot as plt
import numpy as np
import argparse
import cv2
```

```python
path = '/content/drive/MyDrive/Plant/Leaf/q_aff_cerris/R_0Class1 (10).jpg'
image = cv2.imread(path)

plt.title('Original')
plt.imshow(image)
plt.show()
```

```python
(B, G, R) = cv2.split(image)

plt.imshow(R)
plt.title('Red')
```

```python
plt.imshow(G)
plt.title('Green')
```

```python
plt.imshow(B)
plt.title('Blue')
```

```python
merged = cv2.merge([B, G, R])
plt.imshow(merged)
plt.title('Merged')
```

```python
zeros = np.zeros(image.shape[:2], dtype="uint8")
plt.imshow(cv2.merge([B, zeros, zeros]))
plt.title('Red')
```

```python
plt.imshow(cv2.merge([zeros, G, zeros]))
plt.title('Green')
```

```python
# visualize each channel in color
plt.imshow(cv2.merge([zeros, zeros, R]))
plt.title('Blue')
```
