```python
# import the necessary packages
from matplotlib import pyplot as plt
import numpy as np
import argparse
import cv2
```

```python
path = '/content/drive/MyDrive/Plant/Leaf/q_aff_cerris/R_0Class1 (10).jpg'
image = cv2.imread(path)
plt.title('Original')
plt.imshow(image)
plt.show()
```

```python
mask = np.zeros(image.shape[:2], dtype='uint8')
cv2.rectangle(mask, (0, 90), (200, 150), 255, -1)
plt.imshow(mask)
plt.title('Rectangular Mask')
```

```python
masked = cv2.bitwise_and(image, image, mask=mask)
plt.imshow(masked)
plt.title('Mask Applied To')
```
